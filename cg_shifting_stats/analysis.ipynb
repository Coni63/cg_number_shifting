```python
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3
```

```python
db = r"../cg_shifting_runner/results.db"

conn = sqlite3.connect(db)
df = pd.read_sql_query("SELECT level_number, level_input, solution FROM results", conn)
conn.close()
```

```python
df
```

```python
def count_sign(solution):
    res = {
        "positive": 0,
        "negative": 0,
    }
    if solution is None:
        return res

    for row in solution.splitlines():
        if row.strip().endswith("+"):
            res["positive"] += 1
        else:
            res["negative"] += 1
    return res

df[["positive", "negative"]] = df["solution"].apply(count_sign).apply(pd.Series)
df[["width", "height"]] = df["level_input"].apply(lambda x: tuple(map(int, x.splitlines()[0].split()))).apply(pd.Series)
```

```python
df
```

```python
df["board_area"] = df["width"] * df["height"]
df["total_cells"] = df["positive"] + df["negative"] + 1
df['ratio'] = df['negative'] / (df['positive'] + df['negative'])
df["fullness"] = df["total_cells"] / df["board_area"]
```

```python
plt.hist(df['ratio'], bins=20)
plt.show
```

```python
plt.figure(figsize=(20, 6))
plt.plot(df['level_number'], df['ratio'], 'o')
plt.vlines(x=300, ymin=0, ymax=1, colors='lightgray', linestyles='dashed')
```
